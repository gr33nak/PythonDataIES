### 1a. Create a function requesting data from sreality

```python
base_url = 'https://www.sreality.cz/api/cs/v2/estates?category_main_cb=1&category_type_cb=1&locality_region_id=10&per_page60&page={}'.format(i)

r = requests.get(base_url)

```

* function should parametrize: 
    * `category_main_cb` - `{'flat':1, 'house':2, 'land':3 }`
    * `category_type_cb` - `{'sell':1,'rent':2}`
    * `locality_region_id` - `{'Praha':10,'Brno':14}`
    * `page` parameter
* use string inputs for `category_main_cb` and `category_type_cb`
* test the validity of inputs
* include try/except clause to handle errors
* function should return JSON data in python types
* do not forget to sleep each request at least 0.5s

````python
def sreality(page, main = "flat", type = "sell", region = "Praha"):
  
    category_main = {'flat':1, 'house':2, 'land':3 }
    category_type = {'sell':1,'rent':2}
    locality_region = {'Praha':10,'Brno':14}
    
    base_url = 'https://www.sreality.cz/api/cs/v2/estates?category_main_cb{mains}&category_type_cb{types}&locality_region_id{regions}&per_page60&page={page}'
    r = requests.get(base_url)
    try:
        url = base_url.format(
            mains = category_main[main],
            types = category_type[type],
            regions = locality_region[region],
            page = page
        )
        r = requests.get(url)
        return r.json()
    except:
        print("Not possible")
d = sreality(0)
d.keys()
````

### 1b. Create a function converting sreality json data into pandas dataframe

````python
def to_df(json_idk):
    return pd.DataFrame(json_idk["_embedded"]['estates'])
to_df(d)
````

### 1c. link function `1b` into function `1a`

````python
def sreality_to_df(page, main = 'flat', type = 'sell', region = 'Praha'):
  
    category_main = {'flat':1, 'house':2, 'land':3 }
    category_type = {'sell':1,'rent':2}
    locality_region = {'Praha':10,'Brno':14}
    
    base_url = 'https://www.sreality.cz/api/cs/v2/estates?category_main_cb={mains}&category_type_cb={types}&locality_region_id={regions}&per_page60&page={page}'
    r = requests.get(base_url)
    try:
        url = base_url.format(
            mains = category_main[main],
            types = category_type[type],
            regions = locality_region[region],
            page = page
        )
        r = requests.get(url)
        d = r.json()
        return to_df(d)
    except:
        print("Not possible")
df = sreality_to_df(0)
df.keys()
````

### 1c. Combining multiple requests into single df

* Function should parametrize:
    * `start_page` and `end_page`
    * request parameters
* construct a list of individual request dfs
* then feed it into `pd.concat` function

````python
def multiple_sreality(start_page, end_page, main = 'flat', type = 'sell', region = 'Praha'):
    for i in range(start_page, end_page):
        return pd.concat([
            sreality_to_df(i, main=main, type=type, region=region)
        ])
data = multiple_sreality(0,4)
data.head()
````

### 2b: GPS

### 2b. Get flat type from name

````python
clean["flat_type"]=clean.name.apply(lambda name:name.split()[2])
clean.head()
````

### 2c. Get area from name
* Naive: select the word before last word
* Then try navigating using the index of `'m²'`
* if this also fail, then you will need to use regex

````python
def area(nm):
    splitted = nm.split()
    index = splitted.index("m²")
    return int(splitted[index-1])
clean["area"]=clean.name.apply(area)
clean
````

### 4b. Test existence of label `cellar` for offers
* again deal with nested list of list structure
* write generic function `test_existence_of_label(offer_labels,label)`

````python
def test_existence_of_label(offer_labels,label):
    return label in [item for sublist in offer_labels for item in sublist]
clean.labelsAll.apply(lambda offer_labels: test_existence_of_label(offer_labels, 'cellar'))
````

### 4c. Test existence of all possible labels
* use apply returning series with all labels

````python
def existence_of_all_labels(offer_labels, labels):
    return pd.Series({
        label:test_existence_of_label(offer_labels,label)
        for label in labels
    })

clean.labelsAll.apply(lambda offer_labels: existence_of_all_labels(offer_labels, labels))
````
